```python
# E = entrances
# S = students
# G = guards  
E, S, G = list(map(int, input().split()))

students = list(map(int, input().split()))

guards = 1
for i in range(1, len(students)):
    if (students[i] != students[i-1]) and (students[i] not in students[:i]):  
        if students[i-1] in students[i+1:]:
            guards += 1
if guards <= G:
    print('Yes.')
else:
    print('No.')
```
